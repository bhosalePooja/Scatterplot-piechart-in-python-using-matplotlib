#Matplotlib is the whole package; and matplotlib.pyplot is a module in matplotlib.Pyplot provides the state-machine interface to the underlying plotting library in matplotlib. This means that figures and axes are implicitly and automatically created to achieve the desired plot.# 

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5,6,7,8]
Y = [5,7,2,8,6,7,3,4]

plt.scatter(X,Y, label='relation', color= 'k')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('scatterplot in python')
plt.legend()
plt.show()
```

```python
X = [1,2,3,4,5,6,7,8]
Y = [5,7,2,8,6,7,3,4]

plt.scatter(X,Y, label='relation', color= 'b', s=200, marker="*")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('scatterplot in python')
plt.legend()
plt.show()
```

#Notes-
#plt.xlabel('')- adding label to x axis
#plt.ylabel('')- adding label t y axis
#plt.title('')- adding title to the chart
#plt.legend()- showing legend Of the chart(i.e. describes about particular symbol/data being rendered)
#plt.show()- shows the plot
#explode- pull out particular slice from pie chart(for the slices you don't want to explode just mention 0)
#radius- resize the piechart using radius
#autopct- adding percentages automatically to each slice(0.1 is for one decimal place, 0.2 is for two decimal places and like that)
#startangle- to mention the starting position of the slices/starting angle in the pie chart and after that remaining slices will be placed counterclockwise.

```python
slices = [7,3,8,2,1,4]
activities = ['sleeping', 'eating', 'working', 'traveling', 'exercise','other']
plt.pie(slices, labels=activities)
plt.title('piechart in python\ncheck it out')
#plt.legend()
plt.show()
```

```python
slices = [7,3,8,2,1,4]
activities = ['sleeping', 'eating', 'working', 'traveling', 'exercise','other']
cols = ['c','b','g','y','r','m']
plt.pie(slices, labels=activities, colors=cols, startangle=90, explode=(0.1,0,0,0,0,0), shadow=True, radius=1.5)
plt.title('piechart in python')
plt.axis('equal')
#plt.legend()
plt.show()
```

```python
slices = [7,3,8,2,1,4]
activities = ['sleeping', 'eating', 'working', 'traveling', 'exercise','other']
cols = ['c','b','g','y','r','m']
plt.pie(slices, 
        labels=activities, 
        colors=cols, 
        startangle=90, 
        explode=(0.1,0,0,0,0,0), 
        radius=1.2,
        autopct='%0.1f%%')
plt.title('piechart in python\ncheck it out')
#plt.legend()
plt.show()
```

```python
#pie chart on dataframe#
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
df = pd.DataFrame([{'Device': 'Desktop', 'Sessions': 10690313},
                         {'Device': 'Mobile', 'Sessions': 16498696},
                         {'Device': 'Tablet', 'Sessions': 1913828}])
df
```

```python

df.plot(kind='pie', y='Sessions', autopct='%1.1f%%', 
        startangle=90, shadow=False, labels=df['Device'], legend = False, fontsize=14, radius=1.2)
```
